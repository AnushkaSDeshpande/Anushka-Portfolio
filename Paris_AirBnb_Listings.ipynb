```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import plotly.graph_objects as go
import plotly.offline as py
```

```python
listings = pd.read_csv("Listings.csv", low_memory=False,encoding='ISO-8859-1')
```

```python
listings.head()
```

```python
listings.info()
```

```python
listings['host_since']=pd.to_datetime(listings['host_since'])
```

```python
listings.info()
```

```python
paris_listings=listings.query('city == "Paris"')
```

```python
paris_listings.head()
```

```python
paris_listings.info()
```

```python
paris_listings=paris_listings.loc[:,['host_since','neighbourhood','city','accommodates','price']]
```

```python
paris_listings.info()
```

```python
paris_listings.isnull().sum()
```

```python
paris_listings.dropna(inplace=True)
```

```python
paris_listings.isnull().sum()
```

```python
paris_listings.duplicated().sum()
```

```python
paris_listings.drop_duplicates(inplace=True)
```

```python
paris_listings.duplicated().sum()
```

```python
paris_listings.reset_index(inplace=True)
```

```python
paris_listings.describe()
```

```python
paris_listings_neighbourhood=(paris_listings
                              .groupby('neighbourhood')
                              .agg({'price':'mean'})
                              .sort_values('price')
                              .reset_index()
)
```

```python
paris_listings_neighbourhood.tail()
```

```python
paris_listing_accommodates=(paris_listings
                            .query('neighbourhood == "Elysee"')
                            .groupby('accommodates')
                            .agg({'price':'mean'})
                            .sort_values('accommodates')
                            .reset_index()
)
```

```python
paris_listing_accommodates.tail()
```

```python
paris_listings.set_index('host_since',inplace=True)
```

```python
paris_listings_over_time=(paris_listings
                          .resample('Y')
                          .agg({'price':'mean',
                                'neighbourhood':'count'})
                          .reset_index()
)
```

```python
paris_listings_over_time['neighbourhood'].rename('New Hosts',inplace=True)
paris_listings_over_time
```

```python
plt.figure(figsize=(7,5))
sns.barplot(paris_listings_neighbourhood,y='neighbourhood',x='price',color='slateblue')
plt.xlabel('Price per Neight(Euros)')
plt.ylabel('Neighbourhood')
plt.title('Average Listing Price by Paris Neighbourhood')
sns.despine()
plt.show()
```

```python
plt.figure(figsize=(7,5))
sns.barplot(paris_listing_accommodates,y='accommodates',x='price',color='slateblue',orient='horizontal')
plt.xlabel('Price per Neight(Euros)')
plt.ylabel('Accomodation Capacity')
plt.title('Average Listing Price by Accommodation Number')
plt.xticks(rotation=90)
sns.despine()
plt.show()
```

```python
plt.figure(figsize=(7,5))
sns.lineplot(paris_listings_over_time,x='host_since',y='neighbourhood')



sns.despine()

plt.show()
```

```python
sns.lineplot(paris_listings_over_time,x='host_since',y='price')

sns.despine()
plt.show()
```

```python
fig , ax=plt.subplots()

ax.plot(
    paris_listings_over_time['host_since'],
    paris_listings_over_time['neighbourhood'],
    label='New Hosts'
    )
ax.set_ylabel('New Hosts')
ax.set_xlabel('Year')

ax2=ax.twinx()
ax2.plot(
    paris_listings_over_time['host_since'],
    paris_listings_over_time['price'],
    label='Average Price',
    color='red'
    )
ax2.set_ylabel('Average Price')
ax.set_ylim(0)
ax2.set_ylim(0)
```
